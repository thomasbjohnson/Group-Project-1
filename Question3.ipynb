```python
#convert duration to minutes
year["Minute"]= (year["duration_ms"]/1000)/60
#divide per category
#no_pop= year.loc[year["Popularity"]== "Not Popular", :]
#no_pop= no_pop["Minute"]
year

#plt.show()
```

```python
# Categorize age by generations
pop_song=[]    
for x in year["popularity"]:
    if  x< 20:
        name =  "Not popular" 
        pop_song.append(name) 
    elif x >= 20 and x < 50:
        name_2 = "Little popular"
        pop_song.append(name_2)
    elif x >= 50 and x < 75:
        name_3= "Popular"
        pop_song.append(name_3)
    elif x >= 75 and x< 100 :
        name_4= "Very popular" 
        pop_song.append(name_4)   

year["Popularity"]= pop_song
```

```python
#box plot
no_pop= year.loc[year["Popularity"]== "Not popular", :]
no_pop1= no_pop["Minute"]
lil_pop= year.loc[year["Popularity"]== "Little popular", :]
lil_pop1= lil_pop["Minute"]
popu= year.loc[year["Popularity"]== "Popular", :]
popu1= popu["Minute"]
very_pop= year.loc[year["Popularity"]== "Very popular", :]
very_pop1= very_pop['Minute']
#create boxplot
data_to_plot=[no_pop1, lil_pop1, popu1, very_pop1]
names=["Not Popular", "A little Popular", "Popular", "Very Popular"]
fig1, ax1 = plt.subplots(figsize=(15, 10))
ax1.set_title('song duration per popularity category',fontsize =25)
ax1.set_ylabel('song duration',fontsize = 14)
ax1.set_xlabel('Category',fontsize = 14)
ax1.boxplot(data_to_plot, labels=names, widths = 0.4, patch_artist=True,vert=True)

plt.ylim(0, 30)


```

```python
#US dat per age
us_data= age[age["country"] == "US"]
us_file= us_data.dropna(how="any")
us_file= us_file[(us_file["age"] > 12) & (us_file["age"] < 90)]
coun_per_age= us_file.groupby(["age"])["playcount"].sum()
us_file
```

```python
types=[]
# Categorize age by generations
    
for x in us_file["age"]:
    if  x< 21:
        name =  "Gen Z" 
        types.append(name) 
    elif x >= 21 and x < 40:
        name_2 = "Gen Y"
        types.append(name_2)
    elif x >= 40 and x < 56:
        name_3= "Gen X"
        types.append(name_3)
    elif x >= 56 :
        name_4= "Baby Boomers" 
        types.append(name_4)   


        
  
us_file["Category"]= types
```

```python
#rename gender
gender=[]
# Categorize age by generations
    
for x in us_file["gender"]:
    if  x == "m":
        name =  "male" 
        gender.append(name) 
    elif x == "f":
        name_2 = "female"
        gender.append(name_2)
    elif x == "n":
        name_3= "non-gender"
        gender.append(name_3)
      

    
        
  
us_file["Gender"]= gender


```
