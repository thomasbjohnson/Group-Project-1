```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
artist_file = "top10s.csv"
```

```python
# Read our Data file with the pandas library
artist_file_df = pd.read_csv(artist_file, encoding="cp1252")
# Show just the header
artist_file_df.head()
```

```python
# Show just the header
artist_file_df.head()
```

```python
# Show mulitple specific columns--note the extra brackets
artist_file_df[["artist", "year"]].head()
```

```python
artist_count=artist_file_df["artist"].value_counts()
#add a condition if the name of the artist is >= 5 then plot on the graph
```

```python
# Create a bar chart based off of the group series from before
count_chart = artist_count.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Artist Name")
count_chart.set_ylabel("Number of Times on Billboard")


plt.show()
plt.tight_layout()
```

```python
artist_cnt= artist_file_df[["artist"]].value_counts()
print(artist_cnt)
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(artist_file_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, artist_cnt, color='b', alpha=0.5, align="edge")
plt.xticks(tick_locations, artist_file_df["artist"], rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(artist_cnt)+4)
```

```python
# Give our chart some labels and a tile
plt.title("Popularity of Artist from 2010-2019")
plt.xlabel("Artist")
plt.ylabel("Number of times charted on Billboard")
```

```python
# Save our graph and show the grap
plt.tight_layout()
#plt.savefig("../")
plt.show()
```
